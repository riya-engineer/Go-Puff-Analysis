## Donut Chart

```python
import matplotlib.pyplot as plt

sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90,
    # Adding wedgeprops creates the donut "hole"
    wedgeprops={'width': 0.6, 'edgecolor': 'green'}
    
)
plt.title('Sales by Fat Content')
plt.axis('equal')  # Ensures pie chart is a circle
plt.show()
```

### Total sales by Item Type

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group and sort sales by item type
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Set figure size
plt.figure(figsize=(10, 6))

# Create bar chart
bar = plt.bar(sales_by_type.index, sales_by_type.values, color='brown')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
# Add value labels on top of each bar
for bar in bar:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

## Funnel chart

```python
# --- Aggregate and Sort Data ---
funnel_data = df.groupby('Outlet Type')['Sales'].sum().sort_values(ascending=False).reset_index()
labels = funnel_data['Outlet Type']
values = funnel_data['Sales']

# Calculate percentages for labeling
total_sales = values.sum()
percent_values = (values / values.iloc[0]) * 100 # Percentage of first stage

# --- Plotting ---
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars, centering them by shifting start points
for i in range(len(labels)):
    width = values[i]
    # Center the bar
    offset = (max(values) - width) / 2
    ax.barh(labels[i], width, left=offset, color=plt.cm.viridis(i/len(labels)), edgecolor='black')
    
    # Add data labels
    ax.text(max(values)/2, i, f'{labels[i]}: ${width:,.0f} ({width/total_sales:.1%})', 
            ha='center', va='center', color='white', fontweight='bold', fontsize=10)

# Aesthetics
ax.set_title('Funnel Chart: Sales by Outlet Type', fontsize=16)
ax.set_xticks([]) # Remove x-axis
ax.invert_yaxis() # Highest value at the top
plt.box(False) # Remove border

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Group by both categories and sum the sales
# Assumes column names: 'Outlet Location Type', 'Item Fat Content', 'Sales'
stacked_data = df.groupby(['Outlet Type', 'Outlet Size'])['Sales'].sum()

# 2. Unstack the data to pivot 'Item Fat Content' into columns
stacked_data = stacked_data.unstack('Outlet Size')

# 3. Plot the stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(10,6))

# 4. Customize the plot
plt.title('Outlet Type and Outlet Size')
plt.ylabel('Total Sales')
plt.xlabel('OutletType')
plt.xticks(rotation=90)
plt.legend(title='Outlet Type') # Optional: Adds title to legend
plt.tight_layout() # Ensures labels don't get cut off
plt.show()


```
